```python
import pandas as pd
import numpy as np

# 1. Load the dataset
df = pd.read_csv("/content/hotel_bookings 2.csv")

# 2. Check for missing values
print(df.isnull().sum())

# 3. Handle missing values
# 'country' missing values can be filled with 'Unknown'
df['country'] = df['country'].fillna('Unknown')
# 'children' missing values can be filled with 0
df['children'] = df['children'].fillna(0)
# 'agent' and 'company' can be filled with 0 (meaning no agent/company)
df['agent'] = df['agent'].fillna(0)
df['company'] = df['company'].fillna(0)

# 4. Feature Engineering (Add useful columns for SQL/Power BI)
# Combine weekend and week nights to get total stay duration
df['total_stays'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

# Combine adults, children, and babies to get total guests
df['total_guests'] = df['adults'] + df['children'] + df['babies']

# Filter out rows where total guests are 0 (invalid bookings)
df = df[df['total_guests'] > 0]

# 5. Export Cleaned Dataset for SQL & Power BI
df.to_csv("hotel_bookings_cleaned.csv", index=False)
print("Data cleaned and saved successfully! Total records:", len(df))
```
